# Correlations

## 1. Introduction

https://support.minitab.com/en-us/minitab-express/1/help-and-how-to/modeling-statistics/regression/supporting-topics/basics/a-comparison-of-the-pearson-and-spearman-correlation-methods/

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vehicles = pd.read_csv('../../week_05/pandas_import_export/data/vehicles.csv')
vehicles.head()
```

## 2. Correlation Analysis

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'])
```

```python
vehicles['City MPG'].corr(vehicles['City MPG'])
```

```python
vehicles.corr()
```

## 3. Plots

### 3.1 Pandas

```python
plt.scatter(vehicles['City MPG'], vehicles['Highway MPG']);
```

```python
vehicles.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
vehicles.corr(method='spearman').style.background_gradient(cmap='coolwarm').set_precision(2)
```

### 3.2 Seaborn

```python
import seaborn as sns
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1, 1, 1)

sns.heatmap(vehicles.corr(method='spearman'), annot=True, fmt='.2f', ax=ax);
```
